## 4. Modeling

```python
f"ukuran data = {data.shape}"
```

pada data train dilakukan pembagian data test dan train. dimana data test akan digunakan sebagai pengecekan performansi diakhir training. ukuran data test sebesar 20 % dari total data train

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=1)
```

```python
train_data.head()
```

```python
X_train = train_data.drop(["y"], axis = 1)
```

```python
y_train = train_data[["y"]]
```

```python
X_test = test_data.drop(["y"], axis = 1)
```

```python
y_test = test_data[["y"]]
```

```python
len(X_train.iloc[0])
```

## 5. Create model using Neural Network

model machine learning menggunakan model Neural Network. Jumlah layer yang digunakan sebanyak 8 layer dimana terdiri dari 1 input laye, 6 hidden layer dan 1 output layer. 1 input layer berupa flatten layer yang memastikan input data memiliki dimensi / jumlah data sebanyak 125, hidden layer memiliki 2 jenis activation function yakni ELU dan RELU kombinasi ini dilakukan untuk meningkatkan performansi secara quality dan kecepatan. adapun dilakukan kombinasi dengan memberikan dropout/membuang beberapa neuron dengan probability 20 % untuk meningkatan performansi model.

```python
model5 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(125,)),
    tf.keras.layers.Dense(500, activation='elu'),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(500, activation='elu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(500, activation='elu'),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```python
model5.summary()
```

```python
# optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01)

model5.compile(optimizer = "adam",
             loss = 'binary_crossentropy',
             metrics=['accuracy'])
```

training dilakukan dengan jumlah epoch atau training sebanyak 10 kali untuk seluruh data yang ada

```python
model5.fit(X_train, y_train, epochs=20)
```

## 6. Evaluation

untuk data test sebanyak 20% dari data training, model memiliki performansi akurasi sebesar 90,6 %

```python
model5.evaluate(X_test, y_test)
```

```python
y_prob = model5.predict(X_test)
y_prob
```

```python
y_pred = y_prob.round()
print(y_pred)
```
