### 1.2 Encode Value

```python
for col in data_train.columns:
    print(col)
    print(data_train[col].unique())
```

```python
def encode_process_auto(data : pd.DataFrame):
    for i in data.columns:
        if data[i].dtype == "object" and len(data[i].unique()) < 50:
            print(i , " = ", data[i].unique(), len(data[i].unique()))
            list_label = data[i].unique()
            mapping = {}
            indeks = 0
            for label in list_label:
                mapping[label] = indeks
                indeks = indeks + 1

            data[i] = data[i].map(mapping)

if __name__ == "__main__":
    encode_process_auto(data_train)
```

### 1.3 split cabin data

```python
data_train[["cabin_1","cabin_2","cabin3"]] = data_train["Cabin"].str.split("/", expand=True)
```

### 1.4 Encode cabin 1 and 3

```python
encode_process_auto(data_train)
```

### 1.5 convert datatype

```python
data_train["cabin_2"] = data_train["cabin_2"].astype(int)
```

### 1.6 drop unnecessary column

### 1.7 Cleaning Test Data

```python
data_test.info()
```

```python
data_test[["cabin_1","cabin_2","cabin3"]] = data_test["Cabin"].str.split("/", expand=True)
encode_process_auto(data_test)
data_test["cabin_2"] = data_test["cabin_2"].astype('Int64')
data_test.drop(columns=["Name", "Cabin"], inplace=True)
```
