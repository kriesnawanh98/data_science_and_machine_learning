# Classification of Skin Cancer

```python
def convert_image_to_array(path = "./dataset/train/benign/", label = "benign/malignant"):
    import os
    import numpy as np
    from PIL import Image
    
    X_train = []
    for filename in os.listdir(path):
        an_image = Image.open(f"{path}{filename}")
        array_image = np.asarray(an_image)
        X_train.append(array_image)
        
    if label == "benign":
        y_train = np.zeros(len(X_train))
    else:
        y_train = np.ones(len(X_train))

    return X_train, y_train
    
if __name__ == '__main__':
    X_train, y_train = convert_image_to_array(path = "./dataset/train/benign/", label = "benign")
    print(len(X_train))
    print(y_train)
```

```python
X_train_benign, y_train_benign = convert_image_to_array(path = "./dataset/train/benign/", label = "benign")
X_train_malignant, y_train_malignant = convert_image_to_array(path = "./dataset/train/malignant/", label = "malignant")
X_test_benign, y_test_benign = convert_image_to_array(path = "./dataset/test/benign/", label = "benign")
X_test_malignant, y_test_malignant = convert_image_to_array(path = "./dataset/test/malignant/", label = "malignant")
```
